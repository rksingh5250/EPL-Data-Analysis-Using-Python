# English Premium League Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
epl_df = pd.read_csv('EPL 20-21.csv')
epl_df.head()
```

xG is expected number of goals in a match

xA is expected number of assists in a match 

```python
epl_df.info()
```

```python
epl_df.describe(include = 'all')
```

Counting number of null values

```python
epl_df.isna().sum()
```

There are no null values in the dataset. Thus we will not drop any row.

```python
#Create 2 new columns
epl_df['MinsPerMatch']= (epl_df['Mins'] / epl_df['Matches']).astype(int)
epl_df['GoalsPerMatch']= (epl_df['Goals'] / epl_df['Matches']).astype(float)
epl_df.head()
```

Counting no. of Penalty goals

```python
total_PenaltyGoals = epl_df['Penalty_Goals'].sum()
print(total_PenaltyGoals)
```

Counting no. of Total goals

```python
total_goals=epl_df['Goals'].sum()
print(total_goals)
```

Counting no. of Penalty Attempts

```python
total_PenaltyAttempts =  epl_df['Penalty_Attempted'].sum()
print(total_PenaltyAttempts)
```

# Pie chart for penalties Attempted vs Missed

```python
plt.figure(figsize=(14,5))
pl_not_scored = epl_df['Penalty_Attempted'].sum()-total_PenaltyGoals
data = [pl_not_scored, total_PenaltyGoals]
lables = ['Missed Penalties', 'Scored Penalties']
color = sns.color_palette('tab10')
plt.pie(data, labels = lables, colors = color, autopct = '%.0f%%')
plt.show()
#autopct is used to display the percentage values
```

Out of all the penalties attempted only 82% we successfully converted to goals.

# Dividing players on basis of age groups

```python
under20 = epl_df[epl_df['Age'] <= 20]
age_20_to_25 = epl_df[(epl_df['Age']>20) & (epl_df['Age']<=25)]
age_25_to_30 = epl_df[(epl_df['Age']>25) & (epl_df['Age']<=30)]
above30 = epl_df[epl_df['Age']>30]
```

```python
x = np.array([under20['Name'].count(), age_20_to_25['Name'].count(), age_25_to_30['Name'].count(), above30['Name'].count()])
mylabels = ["<20", ">20 & <=25", ">25 & <=30", ">30"]
plt.title('Total players with Age Groups', fontsize=25)
plt.pie(x, labels = mylabels, autopct="%.1f%%")
plt.show()
```

# Plotting total assists from each club

```python
Assists_by_clubs = pd.DataFrame(epl_df.groupby('Club', as_index = False)['Assists'].sum())
sns.set_theme(style="whitegrid", color_codes=True)
ax = sns.barplot(x='Club', y='Assists', data= Assists_by_clubs.sort_values(by="Assists"), palette='Set2')
ax.set_xlabel("Club",fontsize=25)
ax.set_ylabel("Assists",fontsize=35)
plt.xticks(rotation=90)
plt.rcParams["figure.figsize"]=(20,8)
plt.title('Plot of Clubs vs Total Assists')
```

# Assist Kings of the league

```python
top_10_assists = epl_df[['Name', 'Club', 'Assists', 'Matches']].nlargest(n=10, columns='Assists')
top_10_assists
```

# Clubs scoring most number of goals

```python
Goals_by_clubs = pd.DataFrame(epl_df.groupby('Club', as_index = False)['Goals'].sum())
sns.set_theme(style="whitegrid", color_codes=True)
ax = sns.barplot(x='Club', y='Goals', data= Goals_by_clubs.sort_values(by="Goals"), palette='rocket')
ax.set_xlabel("Club",fontsize=25)
ax.set_ylabel("Goals",fontsize=35)
plt.xticks(rotation=90)
plt.rcParams["figure.figsize"]=(20,8)
plt.title('Plot of Clubs vs Total Goals')
```

# Pie chart to plot the Goals with and without assist

```python
plt.figure(figsize=(14,7))
assists = epl_df["Assists"].sum()
data = [total_goals - assists, assists]
labels = ['Goals without assists', "Goals with assists"]
color = sns.color_palette("Set1")
plt.pie(data, labels = labels, colors=color, autopct="%.0f%%")
plt.show()
```

# Chart to show 10 players with highest no. of Yellow Cards

```python
epl_yellow = epl_df.sort_values(by='Yellow_Cards',ascending=False)[:10]
plt.figure(figsize=(20,6))
plt.title('Players with most yellow cards')
c = sns.barplot(x=epl_yellow['Name'], y=epl_yellow['Yellow_Cards'], label="Players", color='yellow')
plt.ylabel('Number of Yellow Cards')
c.set_xticklabels(c.get_xticklabels(), rotation=45)
plt.show()
```

# Chart to show 10 players with highest no. of Red Cards

```python
epl_red = epl_df.sort_values(by='Red_Cards',ascending=False)[:10]
plt.figure(figsize=(20,6))
plt.title('Players with most red cards')
c = sns.barplot(x=epl_red['Name'], y=epl_red['Red_Cards'], label="Players", color='red')
plt.ylabel('Number of Red Cards')
c.set_xticklabels(c.get_xticklabels(), rotation=45)
plt.show()
```
